```python
import numpy as np
from sklearn.datasets import load_digits
import pandas as pd
```

```python
my_data = load_digits()
print(my_data)
```

```python
df = pd.DataFrame(my_data.data)
df.head()
```

```python
X = df
print(X)
```

```python
Y = my_data.target
print(Y)
```

```python
scaled_X = (X-X.mean())
print(scaled_X)
```

```python
from numpy.linalg import eig
covariance = scaled_X.cov()
print(covariance)
```

```python
value, vector = eig(covariance)
print(value)
```

```python
print(vector)
```

```python
# Checking the eigen values to be selected
value_sum = sum(value)
print(value_sum)
```

```python
percentage = (value/value_sum)
print(percentage)
```

```python
selected_columns = percentage>percentage.mean()
print(selected_columns)
```

```python
print(scaled_X.loc[:,selected_columns])
```
