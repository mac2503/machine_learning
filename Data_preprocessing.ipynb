```python
import pandas as pd
```

```python
c = pd.read_csv('C:/Users/mac/Desktop/AWCustomers.csv')
c.head()
```

```python
c.shape
```

```python
na = c.isna()
na.sum()
```

```python
c.drop(columns=['Title','MiddleName','Suffix','AddressLine2'], axis=1,inplace = True)
```

```python
c.head()
```

```python
c.drop(columns=['AddressLine1','City','PostalCode','StateProvinceName'], axis=1,inplace = True)
c.head()
```

```python
c.drop(columns=['CustomerID','FirstName','LastName','PhoneNumber','LastUpdated'], axis=1,inplace = True)
c.head()
```

```python
c['BirthDate']= pd.to_datetime(c['BirthDate'])
c['BirthDate'].dtype
```

```python
import datetime
Today = datetime.datetime(2020, 8, 23)
def age(BD,today=Today):
    d=today-BD
    return d.days//365
```

```python
c['Age']=c['BirthDate'].apply(lambda i: age(i))
c.drop(['BirthDate'],axis=1,inplace=True)
c.head()
```

```python
c.nunique()
```

```python
countries = pd.get_dummies(c['CountryRegionName'], drop_first=True)
c.drop(columns=['CountryRegionName'], axis=1, inplace=True)
```

```python
c = pd.concat([countries,c], axis=1)
c.head()
```

```python
c['Education']=c['Education'].map({'Bachelors':1,'Partial College':2,'High School':3,'Graduate Degree':4,'Partial High School':5})
c['Occupation']=c['Occupation'].map({'Clerical':1,'Skilled Manual':2,'Manual':3,'Management':4,'Professional':5})
c.head()
```

```python
c['Gender']=c['Gender'].map({'F':1,'M':2})
c['MaritalStatus']=c['MaritalStatus'].map({'S':1,'M':2})
c.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
s = MinMaxScaler()
```

```python
changed = s.fit_transform(c[['YearlyIncome','Age']])
c['YearlyIncome_scaled']=changed[:,0]
c['Age_scaled']=changed[:,1]
c = c.drop(['YearlyIncome','Age'], axis=1)
c.head()
```

```python
import seaborn as sns
```

```python
correlation = c.corr()
sns.heatmap(correlation, 
        xticklabels=correlation.columns,
        yticklabels=correlation.columns)
```
